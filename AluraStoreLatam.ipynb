#1. Análisis de facturación

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.



```python
import pandas as pd

# Definición de URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Lectura de los archivos CSV
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Cálculo del ingreso total de cada tienda
ingresos = {
    "Tienda 1": tienda1["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

# Mostrar resultados
for tienda, ingreso in ingresos.items():
    print(f"{tienda}: ₡{ingreso:,.2f}")
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda

```python
import matplotlib.pyplot as plt

# Cantidad de productos vendidos por categoría en cada tienda

ventas_tienda1 = tienda1['Categoría del Producto'].value_counts()
ventas_tienda2 = tienda2['Categoría del Producto'].value_counts()
ventas_tienda3 = tienda3['Categoría del Producto'].value_counts()
ventas_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Mostrar resultados
print("Tienda 1 - Categorías más populares:")
print(ventas_tienda1, "\n")

print("Tienda 2 - Categorías más populares:")
print(ventas_tienda2, "\n")

print("Tienda 3 - Categorías más populares:")
print(ventas_tienda3, "\n")

print("Tienda 4 - Categorías más populares:")
print(ventas_tienda4)

# Graficar los resultados
def graficar_ventas(ventas, tienda_nombre):
    df = ventas.reset_index()
    df.columns = ['Categoría', 'Cantidad']
    df.plot(kind='bar', x='Categoría', y='Cantidad', title=f'{tienda_nombre} - Productos Vendidos por Categoría')
    plt.tight_layout()
    plt.show()

graficar_ventas(ventas_tienda1, "Tienda 1")
graficar_ventas(ventas_tienda2, "Tienda 2")
graficar_ventas(ventas_tienda3, "Tienda 3")
graficar_ventas(ventas_tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda

En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.


```python
import pandas as pd

# Cargar los datos desde las URLs
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer los datos
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Verificar nombres de columnas (opcional)
# print(tienda1.columns)

# Calcular el promedio de calificaciones de clientes por tienda
promedios_calificaciones = {
    "Tienda 1": tienda1["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}

# Mostrar los resultados
for tienda, promedio in promedios_calificaciones.items():
    print(f"{tienda}: {promedio:.2f}")

# Preparar datos
tiendas = list(promedios_calificaciones.keys())
promedios = list(promedios_calificaciones.values())

# Crear gráfico
plt.figure(figsize=(8, 5))
bars = plt.bar(tiendas, promedios)

# Agregar etiquetas a las barras
for bar, valor in zip(bars, promedios):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.05, f"{valor:.2f}",
             ha='center', va='bottom')

# Personalizar el gráfico
plt.title("Promedio de Calificaciones por Tienda")
plt.xlabel("Tienda")
plt.xticks(tiendas)
plt.yticks([])  # Eliminar los números del eje Y
plt.ylim(0, 5)  # Suponiendo escala de 0 a 5
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
import pandas as pd
import matplotlib.pyplot as plt

# URLs de datos
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/main/base-de-datos-challenge1-latam/tienda_4.csv"

# Leer datos
tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

# Función para graficar con etiquetas y sin ejes
def graficar_barras(series, titulo, color='blue'):
    plt.figure(figsize=(10, 4))
    bars = plt.bar(series.index, series.values, color=color)
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f"{int(bar.get_height())}",
                 ha='center', va='bottom', fontsize=9)
    plt.title(titulo)
    plt.xticks(rotation=45, ha='right')
    plt.yticks([])
    plt.box(False)
    plt.tight_layout()
    plt.show()

# Función principal por tienda
def mostrar_top_bottom_ventas(tienda_df, nombre_tienda):
    conteo = tienda_df["Producto"].value_counts()
    top = conteo.head(5)
    bottom = conteo.tail(5)

    graficar_barras(top, f'{nombre_tienda} - Top 5 Productos Más Vendidos')
    graficar_barras(bottom, f'{nombre_tienda} - Top 5 Productos Menos Vendidos', color='gray')

# Aplicar a cada tienda
mostrar_top_bottom_ventas(tienda1, "Tienda 1")
mostrar_top_bottom_ventas(tienda2, "Tienda 2")
mostrar_top_bottom_ventas(tienda3, "Tienda 3")
mostrar_top_bottom_ventas(tienda4, "Tienda 4")

```

# 5. Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
import matplotlib.pyplot as plt

# Calcular el costo de envío promedio por tienda
envio_promedio = {
    "Tienda 1": tienda1["Costo de envío"].mean(),
    "Tienda 2": tienda2["Costo de envío"].mean(),
    "Tienda 3": tienda3["Costo de envío"].mean(),
    "Tienda 4": tienda4["Costo de envío"].mean()
}

# Mostrar resultados por tienda
for tienda, costo in envio_promedio.items():
    print(f"{tienda}: ₡{costo:.2f}")

# Visualizar en gráfico de barras
plt.figure(figsize=(8, 4))
bars = plt.bar(envio_promedio.keys(), envio_promedio.values(), color='orange')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f"₡{bar.get_height():,.0f}",
             ha='center', va='bottom')
plt.title("Costo de Envío Promedio por Tienda")
plt.yticks([])
plt.box(False)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

ingresos = {
    "Tienda 1": tienda1["Precio"].sum(),
    "Tienda 2": tienda2["Precio"].sum(),
    "Tienda 3": tienda3["Precio"].sum(),
    "Tienda 4": tienda4["Precio"].sum()
}

plt.figure(figsize=(10, 5))
bars = plt.barh(list(ingresos.keys()), list(ingresos.values()), color='#2a9d8f')
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1000, bar.get_y() + bar.get_height()/2, f"₡{width:,.0f}", va='center')
plt.title("Ingresos Totales por Tienda", fontsize=14)
plt.xlabel("Ingresos en colones", fontsize=12)
plt.xticks([])
plt.box(False)
plt.tight_layout()
plt.show()

categorias = tienda1["Categoría del Producto"].value_counts()
plt.figure(figsize=(7, 7))
plt.pie(categorias, labels=categorias.index, autopct='%1.1f%%', startangle=90,
        colors=plt.cm.Paired.colors, wedgeprops={'edgecolor': 'white'})
plt.title("Distribución de Categorías de Productos - Tienda 1", fontsize=14)
plt.tight_layout()
plt.show()

import pandas as pd

tienda1["Rango de Precio"] = pd.cut(tienda1["Precio"], bins=[0, 50000, 100000, 150000, 200000, float('inf')],
                                    labels=["₡0-50k", "₡50k-100k", "₡100k-150k", "₡150k-200k", "₡200k+"])

media_calificacion = tienda1.groupby("Rango de Precio", observed=True)["Calificación"].mean()

plt.figure(figsize=(8, 5))
bars = plt.bar(media_calificacion.index.astype(str), media_calificacion.values, color='#264653')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.05, f"{height:.2f}", ha='center')
plt.title("Calificación Promedio por Rango de Precio - Tienda 1", fontsize=14)
plt.ylabel("Calificación Promedio", fontsize=12)
plt.ylim(0, 5)
plt.box(False)
plt.tight_layout()
plt.show()
```
